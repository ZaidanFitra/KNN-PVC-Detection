```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('plotting-Test.csv')

# Sampling rate
fs = 267

df['Time_sec'] = df.index / fs

# Rentang waktu yang ingin ditampilkan
start_time = 0  
end_time = 6    

filtered_df = df[(df['Time_sec'] >= start_time) & (df['Time_sec'] <= end_time)].copy()

fig, ax = plt.subplots(figsize=(15, 6))

ax.plot(filtered_df['Time_sec'], filtered_df['ECG'], label='Sinyal EKG', color='black', linewidth=1.4)

ax.set_xlabel('Waktu (s)')
ax.set_ylabel('Amplitudo (mV)')
ax.set_title('Sinyal EKG')

x_minor_ticks = np.arange(start_time - 0.2, end_time + 0.2, 0.04)
x_major_ticks = np.arange(start_time - 0.2, end_time + 0.2, 0.2)
x_label_ticks = np.arange(start_time, end_time + 1, 1.0)

for x in x_minor_ticks:
    ax.axvline(x=x, color='red', linestyle='-', linewidth=0.08, zorder=0)
for x in x_major_ticks:
    ax.axvline(x=x, color='red', linestyle='-', linewidth=0.4, zorder=0)

ax.set_xticks(x_label_ticks)

raw_min = filtered_df['ECG'].min()
raw_max = filtered_df['ECG'].max()

y_min = np.floor(raw_min) - 1
y_max = np.ceil(raw_max) + 1

y_minor_ticks = np.arange(-10, 10.1, 0.1)
y_major_ticks = np.arange(-10, 10.1, 0.5)
y_label_ticks = np.arange(-10, 11, 1.0)

for y in y_minor_ticks:
    ax.axhline(y=y, color='red', linestyle='-', linewidth=0.08, zorder=0)
for y in y_major_ticks:
    ax.axhline(y=y, color='red', linestyle='-', linewidth=0.4, zorder=0)

ax.set_yticks(y_label_ticks)
ax.set_ylim(y_min, y_max)

ax.legend()
plt.tight_layout()
plt.show()
```
